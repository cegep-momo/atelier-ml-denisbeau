```python
# Cell 1: Imports
import csv
import joblib
import os
```

```python
# Cell 2: Function to get Pokémon info from Pokedexx
def rechercheInformationsPokemon(numPokemon, Pokedex):
    infosPokemon = []
    for pokemon in Pokedex:
        if int(pokemon[0]) == numPokemon:
            infosPokemon = [pokemon[0], pokemon[1], pokemon[4], pokemon[5], pokemon[6],
                            pokemon[7], pokemon[8], pokemon[9], pokemon[10]]
            break
    return infosPokemon
```

```python
# Cell 3: Function to predict battle winner
def prediction(numeroPokemon1, numeroPokemon2, Pokedex, model_path):
    pokemon1 = rechercheInformationsPokemon(numeroPokemon1, Pokedex)
    pokemon2 = rechercheInformationsPokemon(numeroPokemon2, Pokedex)
    
    modele_prediction = joblib.load(model_path)
    
    pred1 = modele_prediction.predict([[pokemon1[2], pokemon1[3], pokemon1[4],
                                        pokemon1[5], pokemon1[6], pokemon1[7], pokemon1[8]]])
    pred2 = modele_prediction.predict([[pokemon2[2], pokemon2[3], pokemon2[4],
                                        pokemon2[5], pokemon2[6], pokemon2[7], pokemon2[8]]])
    
    print(f"COMBAT OPPOSANT : ({numeroPokemon1}) {pokemon1[1]} à ({numeroPokemon2}) {pokemon2[1]}")
    print(f" {pokemon1[1]}: {pred1[0]}")
    print(f" {pokemon2[1]}: {pred2[0]}\n")
    
    if pred1 > pred2:
        print(f"{pokemon1[1].upper()} EST LE VAINQUEUR !")
    else:
        print(f"{pokemon2[1].upper()} EST LE VAINQUEUR !")
```

```python
# Cell 4: Load Pokedex and run prediction
script_dir = os.path.dirname(__file__)
pokedex_path = os.path.join(script_dir, 'datas', 'pokedex.csv')
model_path = os.path.join(script_dir, 'modele', 'modele_pokemon.mod')

with open(pokedex_path, newline='') as csvfile:
    pokedex = list(csv.reader(csvfile))
    next(pokedex)  # skip header
    prediction(368, 598, pokedex, model_path)
```
